```python
from obspy.clients.fdsn import Client
from obspy.core import UTCDateTime, Stream
```

Now a line to define that we want data from the IRIS service:

```python
iris = Client("IRIS")
```

Now a couple of lines to set the times in between which we want to look at data. You can change these - the format is (Year, Month, Day, Hour, Minutes, Seconds)

```python
start_time = UTCDateTime(2023, 10, 4, 0, 0, 0) # (YYYY, m, d, H, M, S)
end_time = UTCDateTime(2023, 10, 4, 1, 0, 0) # (YYYY, m, d, H, M, S)
```

Now we can pull data for your event. We need two lines of code to set which network station we'll pull data from. You can change this to a different station - see the list at

https://earthquake.usgs.gov/monitoring/operations/network.php?virtual_network=GSN.

Choose one from that list, and change the network and station codes to match it.

```python
network = 'IU'
station = 'DAV'
```

And now we have two lines to get the data from that network station between those times, and plot it.

```python
trace = iris.get_waveforms(network, station, '00', "LHZ", start_time, end_time)
trace.plot()
```
